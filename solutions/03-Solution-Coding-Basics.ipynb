### Exercise 2

```r
GOOG <- c(1205, 1248, 1263, 1268, 1238)
AMZN <- c(1802, 1829, 1863, 1808, 1817)

days <- c('Mon', 'Tue', 'Wed', 'Thu', 'Fri')

names(GOOG) <- days
names(AMZN) <- days

stocksPrices <- rbind(GOOG, AMZN)
stocksPrices
```

```r
FB <- c(210, 214, 217, 176, 174)
names(FB) <- days

(stocksPrices <- rbind(stocksPrices, FB))
```

```r
(stocksPrices <- cbind(stocksPrices, average = rowMeans(stocksPrices)))
```

```r
percentageGain <- (stocksPrices[,'Fri'] - stocksPrices[,'Mon']) / stocksPrices[,'Mon'] * 100
(stocksPrices <- cbind(stocksPrices, percentageGain = round(percentageGain, 2)))
```
