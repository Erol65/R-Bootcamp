### Exercise 15
The default geom for `stat_summary()` is `geom_pointrange()` (see the `stat`) argument.

But, the default stat for `geom_pointrange()` is `identity()`, so use `geom_pointrange(stat = "summary")`.

```r
ggplot(data = diamonds) +
  geom_pointrange(
    mapping = aes(x = cut, y = depth),
    stat = "summary",
  )
```

The default message says that `stat_summary()` uses the mean and sd to calculate the point, and range of the line. So lets use the previous values of `fun.ymin`, `fun.ymax`, and `fun.y`:

```r
ggplot(data = diamonds) +
  geom_pointrange(
    mapping = aes(x = cut, y = depth),
    stat = "summary",
    fun.ymin = min,
    fun.ymax = max,
    fun.y = median
  )
```

### Exercise 16
The `geom_col()` function has different default than `geom_bar()`. The default stat of `geom_col()` is `identity()` stat. This means that `geom_col()` expects that the data is already preprocessed into xvalues and yvalues representing the bar height.

```r
demo <- tribble(
  ~cut,         ~freq,
  "Fair",       1610,
  "Good",       4906,
  "Very Good",  12082,
  "Premium",    13791,
  "Ideal",      21551
)

ggplot(data = demo) +
  geom_bar(mapping = aes(x = cut, y = freq), stat = "identity")
```

Or if use `geom_col()` we don't need to specify the `stat` since the default is already identity:

```r
ggplot(data = demo) +
  geom_col(mapping = aes(x = cut, y = freq))
```
